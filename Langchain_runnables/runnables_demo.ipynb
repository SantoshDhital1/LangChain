```python
import random
class FakeLLm:

    def __init__(self):
        print("LLM created")

    def predict(self, prompt):
        response_list = [
            'Kathmandu is the capital of Nepal',
            'NPL is the cricket league',
            'AI stands for Artificial Intelligence'
        ]

        return {'response': random.choice(response_list)}
```

```python
llm = FakeLLm()
```

```python
llm.predict('What is the capital of Nepal ?')
```

```python
class FakePromptTemplate:

    def __init__(self, template, input_variable):
        self.template = template
        self.input_variable = input_variable

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template = FakePromptTemplate(
    template= 'write a {length} poem about {topic}',
    input_variable=['length', 'topic']
)
```

```python
prompt = template.format({'length': 'short', 'topic': 'Nepal'})
```

```python
llm.predict(prompt)
```

```python
class FakeLLmChain:

    def __init__(self, llm , prompt):
        self.llm = llm
        self.prompt = prompt

    def run(self, input_dict):

        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)

        return result['response']
```

```python
template = FakePromptTemplate(
    template= 'write a {length} poem about {topic}',
    input_variable=['length', 'topic']
)
```

```python
llm = FakeLLm()
```

```python
chain = FakeLLmChain(llm, template)
```

```python
chain.run({'length': 'short', 'topic': 'nepal'})
```
