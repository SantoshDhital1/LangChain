## Built-in tools -> DuckDuckGoSearch

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tools = DuckDuckGoSearchRun()

results = search_tools.invoke("Who is the prime minister of Nepal ?")

print(results)
```

```python
print(search_tools.name)
print(search_tools.description)
print(search_tools.args)
```

## Built-in tools -> Shell tool

```python
from langchain_community.tools import ShellTool

search_tools = ShellTool()

results = search_tools.invoke('whoami')

print(results)
```
