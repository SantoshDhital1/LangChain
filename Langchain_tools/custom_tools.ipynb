```python
from langchain_core.tools import tool

# step1: Create a function 
def multiply(a, b):
    """Multiply two number"""
    return a*b
```

```python
# step2: Add type hints
def multiply(a: int, b: int) -> int:
    """Multiply two number"""
    return a*b
```

```python
# step3: Add tool decorator
@tool
def multiply(a:int, b:int) -> int:
    """Multiply two number"""
    return a*b
```

```python
results = multiply.invoke({'a': 3, 'b': 7})
print(results)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema())
```

## Using StructuredTool

```python
from langchain_core.tools import StructuredTool
from pydantic import Field, BaseModel
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add.")
    b: int = Field(required=True, description="The second number to add.")
```

```python
def multiply_func(a:int, b:int) -> int:
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name='multiply',
    description='Mutiply two numbers',
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a': 64, 'b': 43})
print(result)
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

## Using BaseTool Class

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
# arg schema using pydantic
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add.")
    b: int = Field(required=True, description="The second number to add.")
```

```python
class MultiplyTool(BaseTool):
    name: str = 'Multiply'
    description: str = 'Multiply two numbers'

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a': 53, 'b': 5})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

## ToolKit

```python
# Custom tools
@tool
def add(a:int, b:int) -> int:
    """Add two number"""
    return a + b

@tool
def multiply(a:int, b:int) -> int:
    """Multiply two number"""
    return a * b
```

```python
class MathToolKit:
    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = MathToolKit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
